```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do sistema
N = 1000  # Comprimento da sequência
P_values = np.linspace(0, 1, 11)  # Valores de potência do ruído a serem testados

# Função para calcular a BER
def calculate_BER(x, w):
    errors = np.sum(x != w)
    return errors / len(x)

# Loop pelos valores de potência do ruído
for P in P_values:
    # Geração da sequência aleatória de bits
    x = np.random.choice([-1, 1], size=N)

    # Ruído gaussiano com média 0 e desvio padrão sqrt(P)
    noise = np.random.normal(0, np.sqrt(P), size=N)

    # Sinal recebido
    y = x + noise

    # Detecção do sinal
    w = np.sign(y)

    # Cálculo da BER
    ber = calculate_BER(x, w)
    print(f"P = {P:.2f} | BER = {ber:.4f}")

    # Plot dos gráficos
    plt.figure()
    plt.subplot(4, 1, 1)
    plt.plot(x)
    plt.title("x[n]")
    plt.subplot(4, 1, 2)
    plt.plot(y)
    plt.title("y[n]")
    plt.subplot(4, 1, 3)
    plt.plot(y)
    plt.title("r[n]")
    plt.subplot(4, 1, 4)
    plt.plot(w)
    plt.title("w[n]")
    plt.tight_layout()
    plt.show()
```
